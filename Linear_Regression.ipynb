```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Make sample data
x = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1)
```

```python
print(x.shape)
print(y.shape)
```

```python
regr = LinearRegression()
regr.fit(x, y)
```

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title(' Lineaer Regression')
plt.scatter(x, y)
plt.plot(x, regr.predict(x) , color= 'orange')
```
